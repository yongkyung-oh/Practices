```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.random(100)*16
y = np.random.random(100)*10
```

```python
plt.scatter(x,y)
```

```python
x_edge = range(0, 16+1, 2)
y_edge = range(0, 10+1, 2)
```

```python
H, xedges, yedges = np.histogram2d(x, y, bins=(x_edge, y_edge))
H = H.T
```

```python
np.sum(H)
```

```python
H_filp = np.flipud(H)
```

```python
plt.imshow(H_filp)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16, 9))
ax1.scatter(x,y)
ax1.grid(True)
ax2.imshow(H_filp,  aspect="auto")
for (j,i),label in np.ndenumerate(H_filp):
    ax2.text(i,j,int(label),ha='center',va='center')
plt.show() 
```
